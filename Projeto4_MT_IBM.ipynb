## Método Responsável por importar e converter moedas.

```python
import requests
import json
```

```python
def conversor_corrente(amount, from_currency, to_currency): #metodo
    # Use uma API de câmbio para obter as taxas de câmbio
    api_url = f"https://api.exchangerate-api.com/v4/latest/{from_currency}"
    requisito = requests.get(api_url)

    if requisito.status_code != 200:
        print(f"Erro ao obter as taxas de câmbio, ou moeda de origem não é suportada: STATUS {requisito.status_code}")
        return None

    data = requisito.json()

    if 'rates' not in data or to_currency not in data['rates']:
        print(f"Moeda de destino '{to_currency}' não é suportada.")
        return None


    exchange_rate = data['rates'][to_currency]
    dia = data['date']
    print("Na data: " + dia)

    # Realização da conversão
    converted_amount = amount * exchange_rate
    return converted_amount
```

```python
#Mostra as moedas disponíveis primeiro
obter_moedas()
print ('Use os códigos de três letras, acima, referente a moeda que deseja:')


while True:
    try:
        amount = float(input("Digite o valor a ser convertido: "))
        break
    except ValueError:
        print("Por favor, insira um valor numérico válido.")

while True:

    from_currency = input("Digite a moeda de origem: ").upper()
    if len(from_currency) == 3:
        break
    else:
        print("Por favor, insira uma moeda de origem válida.")

while True:
    to_currency = input("Digite a moeda de destino: ").upper()
    if len(to_currency) == 3:
        break
    else:
        print("Por favor, insira uma moeda de destino válida.")

result = conversor_corrente(amount, from_currency, to_currency)

if result is not None:
    print(f"{amount} {from_currency} equivalem a {result} {to_currency}")
```
